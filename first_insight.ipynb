**We can see that some columns have only 1 unique value for each row. Thus these columns do not give any differentiative information for memes. Let's filter out columns which have different values:** 

```python
metacols = []
for col in meta_data.columns:
    if len(meta_data[col].unique())>1:
        metacols.append(col)

metacols
```

```python
meta_data[metacols].head()
```

**I suspect that there are similar columns in meta_data table:**

```python
sum(meta_data["og:title"]!=meta_data["twitter:title"]) #0
sum(meta_data["og:image"]!=meta_data["twitter:image"]) #0
sum(meta_data["description"]!=meta_data["twitter:description"]) #16
meta_data[metacols][meta_data["description"]!=meta_data["twitter:description"]] # these are just NaN values
sum(meta_data["description"]!=meta_data["og:description"]) #16
meta_data[metacols][meta_data["description"]!=meta_data["og:description"]] # these are just NaN values

```

**We have only 4 columns with distinctive values in meta_data table:**

```python
meta_data["meta_title"]=meta_data["og:title"]
meta_data["meta_image"]=meta_data["og:image"]
meta_data["meta_desc"]=meta_data["og:description"]
meta_data["meta_url"]=meta_data["og:url"]
newcols = ["meta_title","meta_image","meta_desc","meta_url"]
meta_data[newcols]
```
